```python
import pandas as pd

# 读取Excel文件
df = pd.read_excel('GongWuYuanList_shenyang2024.xlsx',header=0, skiprows=1)
df
```

```python
df_female = df[df['性别'] == '女']
df_female
```

```python
df_female.to_excel('GongWuYuanList_shenyang2024Female.xlsx', index=False)
```

```python
unis = list(df["毕业院校"])
```

```python
unique = []
counter = []
for i in unis:
    if i not in unique:
        unique.append(i)
        counter.append(1)
    else:
        index = unique.index(i)
        counter[index] += 1

print(unique)
print(counter)
```

```python
#sort the unique list by counter
unique_sorted = [x for _, x in sorted(zip(counter, unique), reverse=True)]
counter_sorted = sorted(counter, reverse=True)
print(unique_sorted)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['Arial Unicode MS']
```

```python
# pie chart with 6 most common universities and others
# 提取前五个标签和对应的计数
top5_counter = counter_sorted[:5]
top5_labels = unique_sorted[:5]

# 计算其他标签的计数
other_counter = sum(counter_sorted[5:])

# 将其他标签合并为一个标签
top5_counter.append(other_counter)
top5_labels.append('其他')

# 绘制饼图
plt.figure(figsize=(10,10))
plt.pie(top5_counter, labels=top5_labels, autopct='%1.1f%%')
plt.show()
```
